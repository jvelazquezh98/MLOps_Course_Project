```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/features.csv')
```

```python
df.head()
```

```python
# make a code to drift the data by changing the values with random 10 quantiles of each column
import numpy as np

df_copy = df.copy()

for col in df_copy.select_dtypes(include=['float64', 'int64']).columns:
    quantiles = df_copy[col].quantile([0.1 * i for i in range(1, 6)]).values
    df_copy[col] = df_copy[col].apply(lambda x: quantiles[np.random.randint(0, 5)])
```

```python
df_copy.head()
```

```python
df_copy.sample(40, random_state=42).to_csv('../data/processed/features_drifted.csv', index=False)
```
